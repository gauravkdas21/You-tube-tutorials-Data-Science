```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'datasets/mtcars.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
cont_cols = df.dtypes[df.dtypes == 'float64'].index.tolist()
cont_cols
```

```python
# univariate 
# bivariate 
# multivariate

# univariate - cont - distplot, box plot | discrete/categorical - countplot, barplot
```

```python
sns.distplot(df['mpg'])
```

```python
sns.distplot(df['qsec'])
```

```python
len(cont_cols)
```

```python
plt.rcParams['figure.figsize'] = [20, 5*2] # w, h

count = 1
for i in cont_cols:
    plt.subplot(2,3, count)
    sns.distplot(df[i])
    count += 1
```

```python
plt.rcParams['figure.figsize'] = [20, 5*2] # w, h

count = 1
for i in cont_cols:
    plt.subplot(2,3, count)
    sns.boxplot(df[i])
    count += 1
```
