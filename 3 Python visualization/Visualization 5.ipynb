```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'datasets/titanic_.csv')
df.head()
```

```python
## multivariate analyses
```

```python
df['Survived'] = df['Survived'].map({0:'not survived', 1: 'Survived'})
```

```python
df= df.rename({'Survived':'Y'}, axis = 1)
```

```python
df = df.drop('PassengerId', axis = 1)
```

```python
df.head()
```

```python
# multivariate analyses
```

```python
sns.FacetGrid(df)
```

```python
sns.FacetGrid(df, col='Pclass') #Pclass - 1,2,3
```

```python
sns.FacetGrid(df, row='Y') #not survived, survived
```

```python
sns.FacetGrid(df, row = 'Y', col = 'Pclass')
```

```python
#sns.FacetGrid(df, row = 'Y', col = 'Sex')
```

```python
g = sns.FacetGrid(df, col='Pclass', row='Y')
g.map(sns.distplot, "Age", hist = False)
```

```python
g = sns.FacetGrid(df, col='Pclass', row='Y')
g.map(sns.countplot, "Sex")
```

```python
g = sns.FacetGrid(df, col='Pclass', row='Y')
g.map(sns.barplot, "Sex", "Age")
```

```python
g = sns.FacetGrid(df, col='Embarked', row='Y')
g.map(sns.countplot, "Sex")
```

```python
g = sns.FacetGrid(df, col='Embarked', row='Y')
g.map(sns.barplot, "Sex", "Age")
```
