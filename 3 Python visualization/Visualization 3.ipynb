```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'datasets/iris.csv')
df.head()
```

```python
df = df.drop('Id', axis = 1)
```

```python
df.head()
```

```python
cont_cols = df.dtypes[df.dtypes == 'float64'].index.tolist()
cont_cols
```

```python
obj_col = 'Species'
```

```python
# cont- cat (boxplot)
```

```python
def plot_bar_box_plots(cont_cols, ob_col, df, est = 'default', type_ = 'bar'):
    
    c = 3
    t= len(cont_cols)

    r = t/c
    if int(r) != r:
        r = int(r) + 1
    else:
        r = int(r)

    plt.rcParams["figure.figsize"] = [20, 5*r]

    count = 1

    for i in cont_cols:

        plt.subplot(r,c, count)
        
        if type_ == 'bar':
            if est == 'median':
                sns.barplot(x = ob_col, y = i, data = df, estimator = np.median)
            else:
                sns.barplot(x = ob_col, y = i, data = df)
        else:
            sns.boxplot(x = ob_col, y = i, data = df)
            
        count += 1
```

```python
plot_bar_box_plots(cont_cols, obj_col, df, type_ = 'bar')
```

```python
plot_bar_box_plots(cont_cols, obj_col, df, type_ = 'box')
```
